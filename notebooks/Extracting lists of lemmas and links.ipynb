4) список списков вхождений лемм в каждый текст корпуса
для подсчета tfidf нужен список с именами файлов и список списков лемм, соответсвующих каждому файлу

```python
# функция для списка лемм одного предложения 
def find_sentence_lemmas(body, j):
    lemmas = []    
    for i in range(len(body[j])):
        if body[j][i].tag == 'W':
            lemmas.append(body[j][i].attrib['LEMMA'])
    return lemmas

print((find_sentence_lemmas(ET.parse('/Users/tatianafrolova/Downloads/SynTagRus/2020_Corpus2_0/Malenkii_prints.tgt').getroot()[1], 2)))
print((find_sentence_lemmas(ET.parse('/Users/tatianafrolova/Downloads/SynTagRus/2020_Corpus2_0/Malenkii_prints.tgt').getroot()[1], 3)))
```

```python
# Объединить списки для предложения в список для текста
# append(А) - добавит список А в качестве элемента, extend - расширит обрабатываемый список на элементы списка А
def find_text_lemmas(body):
    text_lemmas = []
    for j in range(len(body)):
       text_lemmas.extend(find_sentence_lemmas(body, j))
    return text_lemmas

print((find_text_lemmas(ET.parse('/Users/tatianafrolova/Downloads/SynTagRus/2020_Corpus2_0/Malenkii_prints.tgt').getroot()[1])))
```

```python
# применим к корпусу
# здесь нужно делать не extend, а append (добавит элемент) и join (сделает из списка строку)
# потому что именно с таким объектом работает TfidfVectorizer
texts_in_corpus_lemmas = []
for f in file_list:
    texts_in_corpus_lemmas.append(' '.join(find_text_lemmas(ET.parse(f).getroot()[1])))
print(len(texts_in_corpus_lemmas))
```

```python
#сохраним в файл
with open('texts_in_corpus_lemmas.txt', 'w') as file: 
    print(texts_in_corpus_lemmas, file = file)
```

```python
#повторяем для отношений
# для предложений
def find_sentence_links(body, j):
    links = []    
    for i in range(len(body[j])):
        if body[j][i].tag == 'W':
            links.append(body[j][i].attrib.get('LINK','root'))
    return links

print((find_sentence_links(ET.parse('/Users/tatianafrolova/Downloads/SynTagRus/2020_Corpus2_0/Malenkii_prints.tgt').getroot()[1], 2)))
print((find_sentence_links(ET.parse('/Users/tatianafrolova/Downloads/SynTagRus/2020_Corpus2_0/Malenkii_prints.tgt').getroot()[1], 3)))
```

```python
# Объединить списки для предложения в список для текста
# append(А) - добавит список А в качестве элемента, extend - расширит обрабатываемый список на элементы списка А
def find_text_links(body):
    text_links = []
    for j in range(len(body)):
       text_links.extend(find_sentence_links(body, j))
    return text_links

print((find_text_links(ET.parse('/Users/tatianafrolova/Downloads/SynTagRus/2020_Corpus2_0/Malenkii_prints.tgt').getroot()[1])))
```

```python
# применим к корпусу
texts_in_corpus_links = []
for f in file_list:
    texts_in_corpus_links.append(' '.join(find_text_links(ET.parse(f).getroot()[1])))
print(len(texts_in_corpus_links))
```
