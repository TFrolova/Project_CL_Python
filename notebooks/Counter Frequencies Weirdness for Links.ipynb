# Frequencies and Weirdness

Идея в том, чтобы то, что сделано для лемм, скопировать для отношений, вдруг что-то вылезет

```python
# сначала откроем списки, которые понадобятся для подсчетов
import ast
with open("prince_links.txt", "r", encoding='utf-8') as file:
    text = file.read()
prince_links = ast.literal_eval(text)

with open("corpus_links.txt", "r", encoding='utf-8') as file:
    text = file.read()
corpus_links = ast.literal_eval(text)
```

```python
# посмотрим на самые частотные слова из "Маленького принца"
from collections import Counter
print(Counter(prince_links).most_common(50))
```

```python
count_prince_links = Counter(prince_links)
type(count_prince_links)
```

```python
len(count_prince_links)
```

```python
# про сортироваку словаря https://www.freecodecamp.org/news/sort-dictionary-by-value-in-python/
sorted_freq = dict(sorted(count_prince_links.items(), key=lambda item: item[1], reverse=True))
sorted_freq
```

```python
# список ключей
absolute_freq = list(sorted_freq.keys())
print(absolute_freq[0:49])
```

```python
print(Counter(corpus_links).most_common(50))
```

```python
count_corpus_links = Counter(corpus_links)
type(count_corpus_links)
```

```python
len(count_corpus_links)
```

```python
# сортировка словаря
sorted_freq_corpus = dict(sorted(count_corpus_links.items(), key=lambda item: item[1], reverse=True))
sorted_freq_corpus
```

```python
# список ключей
absolute_freq_corpus = list(sorted_freq_corpus.keys())
print(absolute_freq_corpus[0:49])
```

```python
type(sorted_freq)
```

```python
#relative_freq_prince
#для каждой пары в словаре sorted_freq нужно  у ключа значение поделить на длину prince_links 
# и записать полученной число в качестве нового значения в новый словарь
relative_freq_prince = {}
for key, value in sorted_freq.items():
    relative_freq_prince[key] = (value / len(prince_links))
```

```python
print(relative_freq_prince)
```

```python
# то же для всего корпуса
relative_freq_corpus = {}
for key, value in sorted_freq_corpus.items():
    relative_freq_corpus[key] = (value / len(corpus_links))
```

```python
print(relative_freq_corpus)
```

```python
weirdness_prince = {}
for key, value in relative_freq_prince.items():
    weirdness_prince[key] = (value / relative_freq_corpus[key])
```

```python
weirdness_prince
```

```python
# нужно отсортировать, чтобы оценить результат
sorted_weirdness = dict(sorted(weirdness_prince.items(), key=lambda item: item[1], reverse=True))
print(sorted_weirdness)
```

```python
new_weirdness_prince = {}
for key, value in relative_freq_prince.items():
    new_weirdness_prince[key] = (value ** 2/ relative_freq_corpus[key])
```

```python
# сортируем это новое изобретение
sorted_new_weirdness = dict(sorted(new_weirdness_prince.items(), key=lambda item: item[1], reverse=True))
print(sorted_new_weirdness)
```
